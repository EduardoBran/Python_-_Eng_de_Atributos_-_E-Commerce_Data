### <br>Criando Dados para os Exercícios

```python
# Dicionário 1
dados1 = {
        'id_disciplina': ['1', '2', '3', '4', '5'],
        'nome': ['Bob', 'Maria', 'Mateus', 'Ivo', 'Gerson'], 
        'sobrenome': ['Anderson', 'Teixeira', 'Amoedo', 'Trindade', 'Vargas']}

# Criando um dataframe a partir do dicionário 1 (adicionando nome das colunas)
df_a = pd.DataFrame(dados1, columns = ['id_disciplina', 'nome', 'sobrenome'])
df_a
```

```python
# # Dicionário 2
dados2 = {
        'id_disciplina': ['4', '5', '6', '7', '8'],
        'nome': ['Roberto', 'Mariana', 'Ana', 'Marcos', 'Maria'], 
        'sobrenome': ['Sampaio', 'Fernandes', 'Arantes', 'Menezes', 'Martins']}

# Criando um dataframe a partir do dicionário 2 (adicionando nome das colunas)
df_b = pd.DataFrame(dados2, columns = ['id_disciplina', 'nome', 'sobrenome'])
df_b
```

```python
# Dicionário 3
dados3 = {
        'id_disciplina': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'id_teste': [81, 75, 75, 71, 76, 84, 95, 61, 57, 90]}

# Criando um dataframe a partir do dicionário 3 (adicionando nome das colunas)
df_c = pd.DataFrame(dados3, columns = ['id_disciplina','id_teste'])
df_c
```

# <br><br>Exercício 1

### - Faça a concatenação dos dataframes a e b pelas linhas e crie um novo dataframe chamado <font color='darkgreen'>df_ab_linha</font>.

```python
# Solução (concatenando pelas linhas)
df_ab_linha = pd.concat([df_a, df_b], ignore_index = True)
df_ab_linha
```

# <br><br>Exercício 2

### - Faça a concatenação dos dataframes a e b pelas colunas e crie um novo dataframe chamado <font color='darkgreen'>df_ab_coluna</font>.

```python
# Solução
df_ab_coluna = pd.concat([df_a, df_b], axis = 1)
df_ab_coluna
```

# <br><br>Exercício 5

### - Faça o outer join entre os dataframes <font color='darkgreen'>df_a</font> e <font color='darkgreen'>df_b</font> usando a coluna <font color='darkgreen'>id_disciplina</font>
(Merge outer join é a união externa completa que produz o conjunto de todos os registros na Tabela A e na Tabela B, com registros correspondentes de ambos os lados, quando disponíveis. Se não houver correspondência, o lado ausente conterá null.)


```python
# Solução
merge_outer_join = pd.merge(df_a, df_b, on = 'id_disciplina', how = 'outer')
merge_outer_join
```
